<h2>Week 5. Structured Data</h2><br>
    <p><b><u>Structured</u></b> data are available in pre-defined and highly-organized formats such as tables, spreadsheets or databases. Machine languages find it easy to understand structured data which are also often termed as quantitative data. Because of the underlying organizational scheme, such data are easy to collect, store, display and query, unlike <b><u>unstructured</u></b> data which are often categorized as qualitative data. The examples of unstructured data include text, video, audio, mobile and social media activity, satellite and surveillance imagery, and the like. More than 80% of data generated and found useful by business are unstructured.</p>
    <h3>Understanding Variables in Datasets</h3><br>
    <p>Data <b>values</b> (or <b>variables</b>) can be <b>quantitative</b> or <b>qualitative</b>. Quantitative variables can be classified as <b><u>discrete</u></b> or <b><u>continuous</u></b>, while qualitative variables include <b><u>categorical</u></b> type of data values.</p>
    <p><b><u>Discrete variables</u></b> are numeric values that have a countable number of values between any two values. For example, the number of students in this class is discrete. We cannot count 12 students and a half.</p>
    <p><b><u>Continuous variables</u></b> are numeric values that have an infinite number of values between any two values. Our height as measured in centimetres,the length of a road as measured in metres or the date and time of receiving a payment are all examples of continuous data. Yet sometimes discrete data can behave as continuous data.</p>
    <p><b><u>Categorical variables</u></b> contain a finite number of categories or distinct groups. Categorical data might not have a logical order. The biological categories of female vs male gender, the type of music such as jazz, classical, metal and other are a few examples of categorical data.</p>

```python
# Import matplotlib's sub-libray .pyplot under alias 'plt'.

import matplotlib.pyplot as plt

# Use magic line to display plots inline.

%matplotlib inline
```

```python
# Histograms are perfect for one variable, best for continuous data.
# Yet histograms may as well suit discrete datasets whose values are almost continuous.

# Let's hand code a list of wine ages.

wine_age =[20, 45, 40, 42, 5, 1, 2, 23, 100]

wine_age
```

```python
# Let's create a generic function that is meant to plot numeric values as a histogram.
# The statement def is used to create new functions.
# Call that function Age and pass the argument a.
# Pass the argument bins and set it to 5 values, which will split the data into 5 intervals called 'bins'.
# Use the .plt object this time.

def Age(x):   
    plt.hist(x, bins=5)
    plt.xlabel('Age')
    plt.title('Distribution by age')
    return

# To see what the defined function produces when called on a specific dataset, 
# call the function by its name 'Age' and set the keyword argument 'x'  to the variable 'wine_age'.
# Function Age() will plot the wine_age list as a histogram

Age(x=wine_age)
```

```python
# Let's hand code one more list that contains the age of employees.

employees_age =[22, 50.5, 45.6, 40.1, 23, 25, 60.2, 32.8, 35, 36.5, 36, 41, 32, 21.9, 39.4, 37]

employees_age

# Note that the length of this list is longer than the list 'wine_age'. In other words, it contains more data points.
# Because the employee data is similar to the wine data, we can recylce Age()by applying it to both data sets  
```

```python
# Let's print out the plot generated by the function Age().

Age(x=employees_age)
```

```python
# Let's read in the xlsx file called 'books.xlsx'. The file should be in your directory.

# In the video material Part 2 of this week, we used the pandas method .read_csv() to import a csv file. 
# But for .xlsx files, use the mehod .read_excel().

import pandas as pd

books = pd.read_excel('books.xlsx', dtype=object)

books

# What does the string 'NaN' stand for?
```

```python
# Remove the empty rows by setting the axis argument to value 0. 

# The argument set to 1 will target columns.

booksClean = books.dropna(axis=0)

booksClean
```

```python
booksClean = booksClean.reset_index(drop=True)

booksClean
```

```python
# DataFrames are very flexible to compute descriptive statistics, which might be useful to explore your data.

# We can count with len() how many items the 'Genre' column contains.
# pandas treat each column as a subset of the entire dataset.
# Set the slicer to the name of a specific column to query & statistically describe the values in that column.

genre=len(booksClean['Genre'])

genre
```

```python
# Yet the pandas module is equipped with many methods to get data summaries and descriptions.
# Get the summary of the data with .describe().

booksClean.describe()
```

```python
# We may also select columns as subsets to be summarized with .describe() method

booksClean.Publisher.describe()

# Try to subset other columns
```

```python
# .info() gives metadata about the number of rows, columns, column data types, memory usage, and the like.

booksClean.info()

# Also try booksClean.head() and booksClean.tail()
```

```python
# Let's recycle the Age() function from above by calling it on the Height column of the booksClean variable.

Age(x=booksClean['Height'])
```

```python
# Write an empty function that is suited for a single purpose of plotting all books by their height.

def books_bar():
    booksClean.plot(kind='bar', title='Books by height', x='Title',legend=False, figsize=(15, 5))
    plt.xlabel('Titles', fontsize=20, color='red')
    plt.ylabel('Height', fontsize=20, color='red')


books_bar()

# Try to set kind to 'barh' or line' or 'area' or 'density'
```

```python
# We can count the number of values per each unique category in the 'Genre' column with value_counts().

No_books = booksClean['Genre'].value_counts()

No_books

# What do we have in the first and second columns of this variable?
# Note that the pandas methods returns descriptive (meta)data as Series or DataFrames

# Also, if you query the columns with numeric data, e.g. Height, you may set the argument bins to a number bins=2
# Binning works best on continuous data, rather than on discrete data, but the Books dataset doesn't have one.
# With continuous data, use histograms to explore the structure and distribution of your data.
```

```python
# Let's fetch labels from the index colum and numbers in the values' column

print(No_books.index)

print(No_books.values)
```

```python
# We'll need the numpy method arange() to organize the data of the 'Genre' column into 5 bars (datasets). 
# So let's import it first.

import numpy as np

# Write the function to count values of unique genre categories and plot those values by unique categories.

def genre_bar():
    
    No_bars = np.arange(5) #returns data as 5 evenly spaced parts
    No_books = booksClean['Genre'].value_counts() # counts the frequencies (values) per category (we did it above)
    plt.bar(No_bars, No_books) #plot 5 bars on x axis and freq on y
    plt.xticks(No_bars, labels=('fiction', 'nonfiction', 'tech', 'science', 'philosophy'), fontsize=10)
    return

print(genre_bar())
```

```python
# You can choose your colors instead of the default ones.

# Define a list of colors.

Colors =['coral', 'lightblue', 'darkgreen', 'navy', 'lightgreen']
# In the defined function below, set the argument color to the Colors list

def genre_bar2():
    
    No_bars = np.arange(5) #returns data as 5 evenly spaced parts
    No_books = booksClean['Genre'].value_counts() # this has been defined in the previous code cells
    plt.bar(No_bars, No_books, color=Colors) #plot 5 bars on x axis and freq on y
    plt.xlabel('Genre', fontsize = 14)
    plt.ylabel('No of Books', fontsize=14)
    plt.xticks(No_bars, labels=No_books.index, fontsize=10)
    plt.title('Distribution of Books by Genre', fontsize=17, color='darkblue')
    return

genre_bar2()
```
