```python
# import first the requests and BeautifulSoup libraries for this coding session

from bs4 import BeautifulSoup as bs
import requests as re
```

```python
# Send a GET request to establish the connection with the webpage

url = re.get("https://www.rottentomatoes.com/critics/latest_reviews") # latest film reviews on R Tomatoes

url
```

```python
# Download the HTML content and bring it into your coding environment

html = url.content

html
```

```python
# Parse the HTML content to be able to traverse the page and find specific elements with the BeautifulSoup methods

html2 = bs(html, 'html5lib')

html2
```

<div class="alert alert-block alert-info">
<p>With the Inspector tool, check the html code of the webpage to identify by which html element and / or attribute you will need to scrape off the content of reviews.</p>

```python
# From the image above, you know which HTML element and attribute you need to pass as arguments to .find_all()

Reviews = html2.find_all('a', class_="critics-latest-reviews__link a")

Reviews
```

```python
# Write a for-loop to extract only the text from the list of HTML code&text created above

Data = []

for i in Reviews:
    Data.append(i.text)

Data
```

```python
# Let's iterate once again through each <a> element to store them in the Data2 variable 
# with a list comprehension construct instead of the for-loop and then save it into an external file

Data2 = [i.text for i in Reviews]

# Instead of the print() command, we'll save scraped data to an external file
# Use the keyword 'with' to write a code block to create a file and save data in it

with open("Reviews.txt", "w", encoding="utf-8")as f:
    f.write(''.join(Data2)) # to save data in a txt file, your data should be a string
    f.close()
```

```python
# Let's wrap the above steps in a user-defined function

def LatestReviews():
    url = re.get("https://www.rottentomatoes.com/critics/latest_reviews")
    html = url.content
    html2 = bs(html, 'html5lib')
    Reviews = html2.find_all('a', class_="critics-latest-reviews__link a")
    Data2 = [i.text for i in Reviews] # a list comprehension construct
    with open("Reviews2.txt", "w", encoding="utf-8")as f: 
        f.write(''.join(Data2))
        f.close()
    return

LatestReviews()
```

```python
# Import the time module to use its functionality in setting up time intervals 
# at which the code will have to be executed automatically

import time

# Note that once you close this file and shut it down in the directory, the code 
# will be interrupted as well as the scraping process
# Once you open this file and run the code next time, the timer will be activated again
# For automatic scraping, it would be best to run a Python script rather than a JN file
```

```python
# Python has special variables and __name__ is one of them
# If we  run our modules (the source files) as main programs, 
# the interpreter will assign the hard-coded string "__main__" to the __name__ variable

__name__
```

```python
# Write the condition which evaluates if this special variable is equal to '__main__'

if __name__ == '__main__':
    while True: # run the forever loop
        LatestReviews() # call the function we wrote above
        wait_time = 10 # set the time variable to 10 to initialize a starting point
        print("Waiting {} minutes".format(wait_time)) # format a reminder
        time.sleep(wait_time * 60) # .sleep() will suspend this program for the time interval indicated

# the default .sleep() time is in sec, e.g. .sleep(10) will hold the code off for 10 sec. 
# You can use floats and integers to indicate time
```
