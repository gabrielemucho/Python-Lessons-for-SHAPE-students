<div class="alert alert-block alert-danger">
    <h3>Processing one text file</h3>
</div>

```python
# Open and then read in the .txt file called 'The Tyger' by William Blake. 


file = open('Tyger.txt', 'r', encoding="utf-8")
Poem = file.read()
print(Poem)
```

```python
# Outside NLTK, we can clean textual data using string methods.

# You may use the .replace() method to clean or to lemmatize your text.
# If you use .replace() for cleaning, the second argument should be a whitespace.

# Remove full stops, commas and ampersands, and turn all text to lower cases with lower()

plainPoem = Poem.replace('.', '').replace(',', '').replace('?', '').replace('&', '').replace('!', '').lower()


# Print our the poem variable to see whether the unwanted items are removed.

print(plainPoem)
```

```python
# Tokenize the poem with split()
# It will return a list of word strings

wordList = plainPoem.split()
print(wordList)
```

```python
# Clean the text: define our own list of stopwords to be removed from the text before we analyze it.

stopWords = ['a', 'the']
   
stopWords
# This time we'll remove 
```

```python
# Start with an empty list in which we will store the cleaned text

wordListClean = []

for w in wordList:
    if w not in stopWords:
        wordListClean.append(w)
        
print(wordListClean)
 
# Look at the poem and decide what other words you find to be insignificant or irrelevant to your analysis.

# Write a new or expanded stoplist and use it to clean this text once again. Print out the cleaned text.
# But you perhaps you want to explore the use of functions words. 
# In poetry, functions words may be used creatively.
```

```python
# Alternatively to the use of for-loop, you may use a list comprehension construct to clean the textual data

wordListClean2 = [w for w in wordList if w not in stopWords]
wordListClean2
```

```python
# The quickest way to count word frequencies is to import the Counter method from the collections module.
# But Counter works with lists. 
# For this reason, we had to convert the string variable 'plainPoem' into the list 'wordList' as above.

from collections import Counter

wordCount = Counter(wordListClean2)
wordCount

# The output though is a dictionary.
```

```python
# You can convert a dictionary into a list with items() methods.
# Use a list comprehension construct

listPoem = [ [k,v] for k, v in wordCount.items() ]

listPoem

# In week 6, we used .most_common() to convert a frequency dictionary into a list with nested tuples
# Here we use .items() to convert a dictionary to a list with nested lists
# It will be easy to convert the resulting data structure into a pandas' DataFrame
```

```python
# Convert the nested list into a pandas' DataFrame
import pandas as pd

df = pd.DataFrame(listPoem, columns = ["Words", "Counts"])

df.index +=1 # to make index default to 1 instead of 0

df
```

```python
# sort the values in the "Counts" column from highest to smallest value.

df.sort_values("Counts", ascending=False)

# From here you may create visuals or otherwise query the dataset
```

```python
# Import the .wordcloud() method from wordcloud and the pypolot sub-library from matplotlib

from wordcloud import WordCloud
import matplotlib.pyplot as plt

# declare the magic line for visuals
%matplotlib inline
```

```python
# Slice off only the words in the Words column, which is a pandas' Series in terms of data type. 
# .generate() method below will NOT take a Series. So you need to convert it to a string.

Text = str(df["Words"].values) # the values property will remove the axes labels and return values only.
```

```python
# Create and generate a word cloud image:
wordcloud = WordCloud().generate(Text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off") # removes the default x and y axes from the visual


# Ideally, the data should have been cleaned and lemmatized or stemmatized before mapping word frequencies.
```

<div class="alert alert-block alert-danger">
    <h3>Processing multiple text files</h3>
</div>

```python
# Before we read in the multiple .txt files, create a list of the files we want to read in 
# We'll work with three .txt files containing three poems by William Blake.
# They are available for download in the folder of video materials under week 7.

filenames = ["Tyger.txt", "TheAngel.txt", "EcchoingGreen.txt"]

filenames
```

```python
# We need to loop through each text item in the filenames list 
# to open, read, clean each text file and extract frequency dictionaries before
# we can convert them into lists and store everything in one empty list called ListFreq

ListFreq = []

for f in filenames:
    a = open(f, "r", encoding="utf-8")
    b = a.read()
    c = b.replace(",", " ").replace("!", " ").replace("?", " ").replace("!", " ").replace("&", " ").lower()
    d = c.split()
    MyStops = ['a', 'the', 'is', 'to', 'their', 'by', 'in', 'of']
    e = [w for w in d if w not in MyStops]
    f = Counter(e) # we imported Count() in a code cell somewhere above
    h = [ [k,v] for k, v in f.items() ]
    ListFreq.append(h) # append all frequency counts to one empty list 

# Here we packed many steps that we applied above to one text into a code block
```

```python
# print out what the variable ListFreq will return

ListFreq

# Note that we got a multidimensional list. In other words, it's a list with other nested lists
# We have 3 text-level dimensions within which there are as many dimensions as there are words in each text
# If we do not flatten the list by one dimension, it's conversion to DataFrame will be problematic
```

```python
# Flatten the multidimensional (nested) list to a list with fewer dimensions

from itertools import chain

FlatList = list(chain.from_iterable(ListFreq))
FlatList

# You can flatten FlatList even more using the same method once more
```

```python
# import pandas as pd # we imported pandas earlier above
# Convert the flattened frequency list to a DataFrame

BlakeFreq = pd.DataFrame(FlatList, columns=["Words", "Count"])

BlakeFreq

# The concept of flattening was introduced in Part 1_Numpy and Arrays, Week 5.
```

```python
text2 = (BlakeFreq["Words"].values) # the .values property will remove the axes labels and return only values

wordcloud2 = WordCloud().generate(str(text2)) # pass th list of items in text2 to the str() function

# Display the generated image:
plt.imshow(wordcloud2, interpolation='bilinear')
plt.axis("off")

# As expected, the most frequent words, featured in the largest font in the visual, are functions words.
# You'll need to clean and lemmatize the words in the Words column first before generating a thematic wordcloud.
```
